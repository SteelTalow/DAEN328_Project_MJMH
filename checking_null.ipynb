```python
import pandas as pd
import os
```

```python
base_dir = os.getcwd()
file_path = os.path.join(base_dir, 'Data', 'taxidata.csv')
```

```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
null_or_empty = df.isnull() | (df == '')

rowwise_missing = null_or_empty.apply(lambda row: list(row.index[row]), axis=1)

rows_with_missing = rowwise_missing[rowwise_missing.apply(len) > 0]

print(rows_with_missing.head(10)) 
```

```python
null_or_empty_per_column = (df.isnull() | (df == '')).sum()

# Display nicely sorted
print(null_or_empty_per_column.sort_values(ascending=False))
```
